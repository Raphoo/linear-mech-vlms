```python
import matplotlib.pyplot as plt

import pandas as pd

dff = pd.read_csv('embeds/single_lr/image_intervention_appendbind_to_noise_loc_0-1_3-1_thisthat_universal_50x_BOTTLE_double_fixed_moved.csv')
dff_flip = pd.read_csv('embeds/single_lr/image_intervention_appendbind_to_noise_loc_3-1_0-1_thisthat_universal_50x_BOTTLE_double_fixed_moved.csv')

query = "Is the potted plant to the left or right of the bottle?"


plt.plot(dff['left_img1_patched'], c='yellowgreen', linestyle='--')
plt.plot(dff['right_img1_patched'], c='yellowgreen', linestyle='-')

plt.plot(dff_flip['left_img1_patched'], c='orchid', linestyle='--')
plt.plot(dff_flip['right_img1_patched'], c='orchid', linestyle='-')

plt.plot(dff['left_img1'], c='blue', linestyle='--')
plt.plot(dff['right_img1'], c='blue', linestyle='-')

w1 = r"LEFT"
w2 = r"RIGHT"


plt.legend([f'"{w1}" prob after Int 1', f'"{w2}" prob after Int 1',\
            f'"{w1}" prob after Int 2', f'"{w2}" prob after int 2',\
            f'"{w1}" prob no intervention', f'"{w2}" prob no internvention',  ],
               loc='upper center',
    bbox_to_anchor=(0.5, -0.15),  # x=0.5 centers it, y=-0.15 moves it below the x-axis
    ncol=2,                        # put legend in two columns
    frameon=False)


plt.xticks(list(range(15)),list(range(5,20)))

plt.ylabel('log likelihood')
plt.xlabel('layer of intervention')
plt.grid()
title_str = f"VLM Beliefs for the Query: '{query}'"

plt.title(title_str)
plt.show()
```
